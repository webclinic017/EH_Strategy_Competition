```python
from pykrx import stock
import pandas as pd
```

```python
def PER_PBR_day_filtering(date, per_min, per_max, mc_group):
    
    kospi_mc = stock.get_market_cap(date, market="KOSPI")
    kosdaq_mc = stock.get_market_cap(date, market="KOSDAQ")
    
    all_mc = pd.concat([kospi_mc,kosdaq_mc], axis=0)[['시가총액']].sort_values('시가총액',ascending=True)
    all_mc['group'] = pd.cut(all_mc.reset_index().index, bins=3, labels=['소형주','중형주','대형주'])
    if mc_group == '소형주':
        filtered_mc = all_mc.groupby("group").get_group("소형주")
    elif mc_group =='중형주':
        filtered_mc = all_mc.groupby("group").get_group("중형주")
    elif mc_group =='대형주':
        filtered_mc = all_mc.groupby("group").get_group("대형주")

    kospi_per_pbr = stock.get_market_fundamental(date, market="KOSPI")
    kosdaq_per_pbr = stock.get_market_fundamental(date, market="KOSDAQ")
    all_per_pbr = pd.concat([kospi_per_pbr ,kosdaq_per_pbr], axis=0)[['PER','PBR']]
    
    filtered_data = pd.merge(left=filtered_mc, right=all_per_pbr, left_index=True, right_index=True).query('PBR != 0')
    cond = (filtered_data['PER'] >= per_min ) & (filtered_data['PER'] <= per_max)
    filtered_data = filtered_data[cond].sort_values('PBR')[0:30]

    filtered_ticker = list(filtered_data.index)
    return filtered_ticker
```
